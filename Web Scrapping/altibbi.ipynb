```python
import requests
from bs4 import BeautifulSoup
import csv
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.chrome.service import Service
```

```python
chrome_options = Options()
chrome_options.add_argument('--headless')
chrome_options.add_argument('--no-sandbox')
chrome_options.add_argument('--disable-dev-shm-usage')
service = Service(executable_path='chromedriver.exe')
options = webdriver.ChromeOptions()
from selenium.webdriver.chrome.service import Service
from selenium import webdriver

service = Service()
driver = webdriver.Chrome(service=service)
```

```python
qna = []
```

```python
def get_href(container):
    question_href = container.contents[1].find('a' , {'class': "question-text"} , href = True).get('href')
    print(question_href)
    try:
        get_qna_main(question_href)
    except:
        print("Connection Timed out While Getting QnA site ...  The process will restart")
#         get_qna(question_href)
    finally:
        print("The second operation done")
```

```python
def get_qna(container):
    question_text = container.contents[1].find('div' , {'class': "question-description-text"}).text.strip()
#     print(question_text)
    answer_text = container.contents[3].find('div' , {'class': "doctor-answer"}).text.strip()
#     print(answer_text)
    qna.append({
        'question' : question_text,
        'answer' : answer_text
    })
```

```python
def get_qna_main(question_href):
    driver.implicitly_wait(30)
    driver.get(question_href)
    soup = BeautifulSoup(driver.page_source, 'lxml')
    container = soup.find_all("section" , {'class' : "question-and-answer-partial-wrapper"})
    get_qna(container[0])
```

```python
def qna_main(pageNumber):
    driver.get("https://altibbi.com/%D8%A7%D8%B3%D8%A6%D9%84%D8%A9-%D8%B7%D8%A8%D9%8A%D8%A9/%D8%A7%D9%85%D8%B1%D8%A7%D8%B6-%D9%86%D9%81%D8%B3%D9%8A%D8%A9?page="+str(pageNumber))
    soup = BeautifulSoup(driver.page_source, 'lxml')
    container = soup.find_all("article" , {'class' : "new-question-item"})
    for i in range(len(container)):
        get_href(container[i])
```

```python
print("Iam in page "+str(2))
qna_main(2)
```

```python
for i in range(1429):
    try:
        print("Iam in page "+str(i+1))
        qna_main(i+1)
    except:
        print("Connection Timed out!! ... The process will restart")
#         qna_main(i+1)
    finally:
        print("The process ended successfully")
```
